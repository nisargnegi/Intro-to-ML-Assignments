```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from hmmlearn import hmm

base_dir = "silver.csv"#"https://www.kaggle.com/datasets/psycon/daily-silver-price-historical-data"

data = pd.read_csv(base_dir)

# Convert the datetime from str to datetime object.
data["Date"] = pd.to_datetime(data["Date"])

# Determine the daily change in gold price.
data["silver_price_change"] = data["Close"].diff()

# Restrict the data to later than 2008 Jan 01.
data = data[data["Date"] >= pd.to_datetime("2008-01-01")]

# Plot the daily gold prices as well as the daily change.
plt.figure(figsize = (15, 10))
plt.subplot(2,1,1)
plt.plot(data["Date"], data["Close"])
plt.xlabel("Date")
plt.ylabel("silver price (usd)")
plt.grid(True)
plt.subplot(2,1,2)
plt.plot(data["Date"], data["silver_price_change"])
plt.xlabel("Date")
plt.ylabel("silver price change (usd)")
plt.grid(True)
plt.show()
data = pd.read_csv(base_dir)

# Convert the datetime from str to datetime object.
data["Date"] = pd.to_datetime(data["Date"])

# Determine the daily change in gold price.
data["silver_price_change"] = data["Close"].diff()

# Restrict the data to later than 2008 Jan 01.
data = data[data["Date"] >= pd.to_datetime("2008-01-01")]

plt.figure(figsize = (15, 10))
plt.subplot(2,1,1)
plt.plot(data["Date"], data["Close"])
plt.xlabel("Date")
plt.ylabel("silver price (usd)")
plt.grid(True)
plt.subplot(2,1,2)
plt.plot(data["Date"], data["silver_price_change"])
plt.xlabel("Date")
plt.ylabel("silver price change (usd)")
plt.grid(True)
plt.show()
```

```python
# Use the daily change in gold price as the observed measurements X.
X = data[["silver_price_change"]].values
# Build the HMM model and fit to the gold price change data.
model = hmm.GaussianHMM(n_components = 3, covariance_type = "diag", n_iter = 50, random_state = 42)
model.fit(X)
# Predict the hidden states corresponding to observed X.
Z = model.predict(X)
states = pd.unique(Z)
```

```python
print("Unique states:")
print(states)
```

```python
print("\nStart probabilities:")
print(model.startprob_)
```

```python
print("\nTransition matrix:")
print(model.transmat_)
```

```python
print("\nGaussian distribution means:")
print(model.means_)
```

```python
print("\nGaussian distribution covariances:")
print(model.covars_)
```

```python
plt.figure(figsize = (15, 10))
plt.subplot(2,1,1)
for i in states:
    want = (Z == i)
    x = data["Date"].iloc[want]
    y = data["Close"].iloc[want]
    plt.plot(x, y, '.')
plt.legend(states, fontsize=16)
plt.grid(True)
plt.xlabel("Date", fontsize=16)
plt.ylabel("silver price (usd)", fontsize=16)
plt.subplot(2,1,2)
for i in states:
    want = (Z == i)
    x = data["Date"].iloc[want]
    y = data["silver_price_change"].iloc[want]
    plt.plot(x, y, '.')
plt.legend(states, fontsize=16)
plt.grid(True)
plt.xlabel("Date", fontsize=16)
plt.ylabel("silver price change (usd)", fontsize=16)
plt.show()
```
